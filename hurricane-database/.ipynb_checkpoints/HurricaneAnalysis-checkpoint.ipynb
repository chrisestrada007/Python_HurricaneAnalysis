```python
# Hurricane Analysis
# -------------------------
# The purpose of this notebook is to create a predictive model for hurricanes.
# Dataset from Kaggle: Hurricanes and Typhoons, 1851-2014
# https://www.kaggle.com/noaa/hurricane-database
```

```python
# Dependancies
import pandas as pd
import csv
import os
import numpy as np

%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
# Import csv
file = "atlantic.csv"
file_df = pd.read_csv(file)
file_df
```

```python
# 1st Look at the data:
# Columns look good
# Date is yyyymmdd
# Each day has 4 time points: 0, 600, 1200, and 1800
file_df.columns
```

```python
# Points per storm
stormPoints = (file_df['ID'].value_counts())
stormPoints.head()

# There have been 1814 recorded storms since 1851

# The most highly recorded were:
# ID        Records
# AL031899    133
# AL141971    118
# AL201969     99
# AL142012     96
# AL041957     95
```

```python
# StormPoints Per Date
stormPointsPerDay = file_df.Date.value_counts()
#stormPointsPerDay.head()

# One week in Sept of 1971 was particularly rainy!
# Date      Records
# 19710912    24
# 19710911    24
# 19710913    21
# 19950827    19
# 19710914    19

stormPointsPerDay = pd.DataFrame(stormPointsPerDay)
stormPointsPerDay = stormPointsPerDay.rename(columns={'Date':'Records'})
stormPointsPerDay['Date'] = stormPointsPerDay.index
stormPointsPerDay = stormPointsPerDay.sort_values("Date")
stormPointsPerDay = stormPointsPerDay.reset_index(drop=True)
columnsTitles = ["Date","Records"]
stormPointsPerDay = stormPointsPerDay.reindex(columns=columnsTitles)
stormPointsPerDay.Date = pd.to_datetime(stormPointsPerDay.Date, format='%Y%m%d', errors='ignore')
stormPointsPerDay
```

```python
# Lets Look at the frequency of storms over time
stormPointsPerDay.plot(x='Date', figsize = (500,10))
#shows Number of storms by day
```
